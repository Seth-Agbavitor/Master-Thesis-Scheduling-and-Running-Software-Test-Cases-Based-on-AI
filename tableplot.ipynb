```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the feature summary data
data = {
    "Feature": ["Failure Rate", "Duration", "Time Since Last Run"],
    "Description": [
        "Percentage of failed test cases",
        "Test case execution time in seconds",
        "Recency of test execution in hours"
    ],
    "Normalization Range": ["0.0 - 1.0", "0.0 - 1.0", "0.0 - 1.0"]
}

# Create DataFrame
df = pd.DataFrame(data)

# Set up the matplotlib figure
plt.figure(figsize=(8, 2))
sns.set(style="whitegrid")

# Plot the table using seaborn's axis
table = plt.table(cellText=df.values,
                  colLabels=df.columns,
                  loc='center',
                  cellLoc='left',
                  colColours=["#83bdf0"] * 3,
                  colWidths=[0.3, 0.5, 0.3])

table.scale(1, 2)
plt.axis('off')
plt.tight_layout()

# Save and show the plot


```
